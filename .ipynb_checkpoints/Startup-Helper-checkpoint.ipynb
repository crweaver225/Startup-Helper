#### Wikipedia
First, I will build a web crawler to get data for all neighborhoods from wikipedia

```python
# Web scrape a wikipedia page for chicago community data
results = requests.get("https://en.wikipedia.org/wiki/Community_areas_in_Chicago")
soup = BeautifulSoup(results.content)
```

```python
# Get neighborhood name, population, size in square miles, and density from wikipedia page
table=soup.find('table')
rows = table.find_all('tr')

table_contents = []

for row in rows[2:-1]:
    
    cell = {}
    
    community_title = row.find('th')
    links = row.find('a', href=True)
    title = links.string
    
    community_data = row.findAll('td')
    population = community_data[1:][0].text
    area = community_data[1:][1].text
    density = community_data[1:][3].text
    
    cell['title'] = title.strip()
    cell['population'] = population
    cell['area_sq_mi'] = area
    cell['density'] = density
    
    table_contents.append(cell)
```

```python
# Convert web scraping to a Dataframe
df=pd.DataFrame(table_contents)
df.head()
```

The first to retrieve all the cafe venues in all those neighborhoods and the second to get more granular metrics about each venue

```python
# Foursquare api info

CLIENT_ID = 'QNY4BLCNST4M2RWWZ5GG3DE44LGK0N5N5ZXU2AB045UZZRMO' # your Foursquare ID
CLIENT_SECRET = 'ATTMHTXYH0COJUK2GQ1IZZ2V00HU2B4MM33P4OSK5HDIL5JL' # your Foursquare Secret
VERSION = '20180605' # Foursquare API version
LIMIT = 1000 # A default Foursquare API limit value
```

```python
# A function that will utilize the Foursquare api to get restaurants for each neighborhood

def getNearbyVenues(names, radius=1000):
    
    venues_list=[]
    found_neighborhoods = []
    for name in names:
 
        url_name = '{}, IL'.format(name)

        url = 'https://api.foursquare.com/v2/venues/search?&client_id={}&client_secret={}&query=cafe&v={}&near={}&radius={}&limit={}&sortByDistance=1'.format(
                CLIENT_ID, 
                CLIENT_SECRET, 
                VERSION, 
                url_name,
                radius, 
                LIMIT)
       
            
        # make the GET request
        json_return = requests.get(url).json()
   
        if len(json_return["response"]) > 0:
              
            if 'venues' in json_return["response"]:
                results = json_return["response"]['venues']
    
                # return only relevant information for each nearby venue
                venues_list.append([(
                    name,  
                    v['name'], 
                    v['id'],
                    v['location']['lat'], 
                    v['location']['lng']) for v in results])
    
        nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
        nearby_venues.columns = ['Neighborhood',  
                      'Venue', 
                      'Venue id',
                      'Venue Latitude', 
                      'Venue Longitude']
    return(nearby_venues)
```

```python
# Load the granular metric data for a specific venue by id using the Foursquare api

def getVisitorDataForVenue(ids):
    
    venue_details_df = pd.DataFrame(columns=['id','tipCount', 'likes', 'rating'])
    
    for id in ids:
       
        url = 'https://api.foursquare.com/v2/venues/{}?&client_id={}&client_secret={}&v={}'.format(
                    id,
                    CLIENT_ID, 
                    CLIENT_SECRET, 
                    VERSION)
        json_return = requests.get(url).json()["response"]
    
        if 'venue' in json_return:
            
            data = {}
        
            venue = json_return['venue']
            
            data['id'] = id
            
            if 'stats' in venue:
                stats = venue['stats']
                if 'tipCount' in stats:
                    data['tipCount'] = stats['tipCount']
            
            if 'likes' in venue:
                likes = venue['likes']
                if 'count' in likes:
                    data['likes'] = likes['count']
            
            if 'rating' in venue:
                data['rating'] = venue['rating']
        
            venue_details_df = venue_details_df.append(data, ignore_index=True)
        
    return venue_details_df
```

### Modeling

We next will try to build out a couple of models that can take in some of the features of each neighborhood and predict the rating for the cafe. Because we have such an uneven distribution of ratings among the neighborhoods, the average rating per neighborhood is not a good indicator of how a new cafe would be rated. Instead, modeling might find interesting combinations of features found in neighborhoods that will better help us understand which neighborhoods a cafe we recieve higher ratings for based on their own feature characterstics. 

Because I am looking for a output from the model that is not a classification, but a continous number, I will look to build some sort of linear model to start. 

```python
train_x = final_venue_df.iloc[:,-8:].values.tolist()
train_y = final_venue_df['rating'].values.tolist()
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(train_x).transform(train_x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, train_y, test_size=0.2, random_state=4)
```

Going to build a linear regression model. 

The results were very poor, with a r score of 0.09

```python
# Building a linear model to see if we can predict ratings for a venue
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
reg.score(X_train, y_train)
```

Next going to try and see if a polynomial linear learner might work better. 

The results were better, but still very poor for the r-score

```python
from sklearn.preprocessing import PolynomialFeatures

pr=PolynomialFeatures(degree=5)
x_pf=pr.fit_transform(train_x).astype(float)
x_pf.shape
```

```python
reg2 = LinearRegression().fit(x_pf, train_y)
reg2.score(x_pf, train_y)
```
